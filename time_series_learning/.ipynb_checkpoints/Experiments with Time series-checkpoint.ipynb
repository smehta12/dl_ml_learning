```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import numpy as np

data = pd.read_csv(r"data_ts.csv")
print("Num observations:%d"%(data.shape[0]))

print(data.head())
```

```python
plt.figure(figsize=(20,7))
plt.plot(data.DateAnalyzed, data.Mean)
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```

```python
# Convert data to time series

ts= data.set_index(pd.DatetimeIndex(data["DateAnalyzed"]))["Mean"]
ts.head()
```

```python
# Lagging the time series by 1 and 2 
ts_lag1_2=pd.concat([ts, ts.shift(1), ts.shift(2)], axis=1)
print(ts_lag1_2.head())
plt.plot(ts, color='blue')
plt.plot(ts.shift(1), color='red')
plt.xticks(rotation=70)
plt.show()

plt.clf()
plt.plot(ts.diff(periods=1), color='blue') # another way to get lag
```

```python
# Window functions for smoothing---ROLLING WINDOW
r=ts.rolling(window=5) # Agrregate the data. Creates the window of every 5 observations. Works like groupby
# print(r.mean().head(7))
# print(r.agg(['sum', 'mean'])).head(7)
ts.plot(color='gray')
# Other functions can also be use like min, max etc.
r.mean().plot(color='red') # Creates mean of every 5 observation
```

```python
# EXPANDING WINDOW

ts.expanding(min_periods=2).mean().plot()
```

```python
# Perform exponentially weigth moving avg.

ts.ewm(span=5, freq='D', min_periods=0, adjust=True).mean().plot(color='red')
ts.rolling(window=5).mean().plot(color='green')
```

```python
# AUTO Correlation

from statsmodels.tsa import stattools
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

stattools.acf(ts) # to find acf for the values
plt.figure(figsize=(20,7))
plot_acf(data.Mean)
plt.show()

# Here the plot doesn't give any good info. Use it after cleaning/de-trending of the data
```

```python
# Real autocorrelation

air_data = pd.read_csv("air_passanger.csv", header=0, parse_dates=[0], names = ['Month', 'Passengers'], index_col = 0)
air_data.plot()

plot_acf(air_data)

acf_result = stattools.acf(air_data)
plt.show()
plt.clf()

plt.subplot(121) 
plt.plot(acf_result)
plt.axhline(y=0,linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(air_data)),linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(air_data)),linestyle='--')
```

```python
# Clean up the data 
#plt.clf()
log_air_passengers = np.log(air_data)
log_air_passengers_diff = log_air_passengers - log_air_passengers.shift()
log_air_passengers_diff.plot()
```

```python
# ACF of clean data
# print(log_air_passengers_diff.head())
# plot_acf(log_air_passengers_diff)
# plt.show()


log_air_passengers_diff = log_air_passengers_diff
log_air_passengers_diff.dropna(inplace = True)

plot_acf(log_air_passengers_diff)
plt.show()

plt.clf()
acf_result = stattools.acf(log_air_passengers_diff)
plt.plot(acf_result)
plt.axhline(y=0,linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(log_air_passengers_diff)),linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(log_air_passengers_diff)),linestyle='--')
plt.show()
```

### ARIMA model developement

```python
# AR code
def ar1(phi = .9, n = 1000, init = 0):
    time_series = [init]
    error = np.random.randn(n)
    for period in range(n):
        time_series.append(error[period] + phi*time_series[-1])
    return pd.Series(time_series[1:], index = range(n))
    
def ar2(phi1 = .9, phi2 = -.8, n = 1000, init = 0):
    time_series = [init, init]
    error = np.random.randn(n)
    for period in range(2,n):
        time_series.append(error[period] + phi1*time_series[-1] + phi2*time_series[-2])
    return pd.Series(time_series[1:], index = range(1,n))
```

```python
# try out different values of phi >=1 as compared to < 1
# sometimes you need to make a large n to see lack of stationarity
a1 = ar1(phi = 1.1, n = 40)
a1.plot()
# Not stationary plot generated
```

```python
# try out different values of phi >=1 as compared to < 1
# sometimes you need to make a large n to see lack of stationarity
a1 = ar1(phi = 1.1, n = 40)
a1.plot()

# No stationary plot generated
```

```python
# ma1 code
def ma1(theta = .5, n = 100):
    time_series = []
    error = np.random.randn(n)
    for period in range(1,n):
        time_series.append(error[period] + theta*error[period-1])
    return pd.Series(time_series[1:], index = range(1,n-1))
```

```python
m1 = ma1(theta = -1000)
m1.plot()

# Generated stationary plot
```

```python
# Model diagnostics
from statsmodels.tsa.stattools import acf, pacf
a1 = ar1(phi = .5, n = 1000)
a1_acf = acf(a1, nlags = 20)
plt.plot(a1_acf)
plt.axhline(y=0,linestyle='--', color = 'black')
plt.axhline(y=-1.96/np.sqrt(len(a1)),linestyle='--', color = 'red')
plt.axhline(y=1.96/np.sqrt(len(a1)),linestyle='--', color = 'red')

# # For AR, PACF matters more
```

```python
# Model diagnostics
a1 = ar1(phi = .5, n = 1000)
a1_pacf = pacf(a1, nlags = 20)
plt.plot(a1_pacf)
plt.axhline(y=0,linestyle='--', color = 'black')
plt.axhline(y=-1.96/np.sqrt(len(a1)),linestyle='--', color = 'red')
plt.axhline(y=1.96/np.sqrt(len(a1)),linestyle='--', color = 'red')

# For AR, PACF matters more. Here, this is an ideal plot. It shows that the value at 0 and 1 has high corelation 
# where as from 2 it is falling. That means it is predicting the value @ 1 properly.
```

```python
# MA acf. # Model diagnostics

m1 = ma1(n = 1000, theta = .9)
m1_acf = acf(m1, nlags = 20)
plt.plot(m1_acf)
plt.axhline(y=0,linestyle='--', color = 'black')
plt.axhline(y=-1.96/np.sqrt(len(m1)),linestyle='--', color = 'red')
plt.axhline(y=1.96/np.sqrt(len(m1)),linestyle='--', color = 'red')

# # For MA, ACF matters more. It shows that where to cutoff the model. i.e. upto what iteration it is provideing 
# highly correlated data.
```

```python
# MA pacf. # Model diagnostics

m1 = ma1(n = 1000, theta = .9)
m1_pacf = pacf(m1, nlags = 20)
plt.plot(m1_pacf)
plt.axhline(y=0,linestyle='--', color = 'black')
plt.axhline(y=-1.96/np.sqrt(len(m1)),linestyle='--', color = 'red')
plt.axhline(y=1.96/np.sqrt(len(m1)),linestyle='--', color = 'red')

# For MA, ACF matters more. 
```
